- 统计满勤（出勤率100%）的天数
要求使用for循环实现。

```python
attendance = [
    [1, 1, 1, 1, 1],  # 第一天，5个学生的出勤情况
    [1, 1, 0, 1, 1],  # 第二天
    [1, 1, 1, 1, 1],  # 第三天
    [1, 0, 1, 1, 1]   # 第四天
]
```

**参考答案**：

```python
attendance = [
    [1, 1, 1, 1, 1],  # 第一天，5个学生的出勤情况
    [1, 1, 0, 1, 1],  # 第二天
    [1, 1, 1, 1, 1],  # 第三天
    [1, 0, 1, 1, 1]   # 第四天
]

perfect_days = 0  # 满勤天数
total_students = len(attendance[0])  # 总学生数

for day, records in enumerate(attendance, 1):
    present = sum(records)  # 当天出勤人数
    rate = present / total_students * 100
    print(f"第{day}天 出勤率：{rate:.0f}%")
    
    if present == total_students:
        perfect_days += 1

print(f"\n满勤天数：{perfect_days}天")
```

- 找出最高和最低月份的销售额
- 计算月平均销售额
要求使用for循环实现，且保留2位小数。

```python
sales = [120.5, 98.7, 145.8, 132.3, 110.5, 128.9, 
        143.2, 156.7, 165.2, 149.8, 138.4, 180.5]
```

**参考答案**：

```python
sales = [120.5, 98.7, 145.8, 132.3, 110.5, 128.9, 
        143.2, 156.7, 165.2, 149.8, 138.4, 180.5]

total = 0
max_sale = sales[0]
min_sale = sales[0]

for sale in sales:
    total += sale
    if sale > max_sale:
        max_sale = sale
    if sale < min_sale:
        min_sale = sale

average = total / len(sales)

print(f"年度总销售额：{total:.2f}万元")
print(f"最高月销售额：{max_sale:.2f}万元")
print(f"最低月销售额：{min_sale:.2f}万元")
print(f"月平均销售额：{average:.2f}万元")
```

**题目**：商品库存进销存统计。给定一组商品编号product_ids和对应的库存数量stock_nums，编写程序：
- 统计库存数量大于100的商品数量
- 计算库存数量的标准差
- 输出库存量最大的三个商品编号
要求使用for循环实现。

```python
product_ids = ['A001', 'A002', 'A003', 'A004', 'A005', 'A006']
stock_nums = [150, 80, 200, 95, 120, 180]
```

**参考答案**：

```python
product_ids = ['A001', 'A002', 'A003', 'A004', 'A005', 'A006']
stock_nums = [150, 80, 200, 95, 120, 180]

# 统计库存>100的商品数量
high_stock_count = 0
for num in stock_nums:
    if num > 100:
        high_stock_count += 1

# 计算标准差
mean = sum(stock_nums) / len(stock_nums)
variance = 0
for num in stock_nums:
    variance += (num - mean) ** 2
std_dev = (variance / len(stock_nums)) ** 0.5

# 获取库存最大的三个商品
stock_info = list(zip(product_ids, stock_nums))
stock_info.sort(key=lambda x: x[1], reverse=True)
top3_products = stock_info[:3]

print(f"库存大于100的商品数量：{high_stock_count}个")
print(f"库存数量的标准差：{std_dev:.2f}")
print("库存量前三的商品：")
for product, num in top3_products:
    print(f"商品{product}: {num}件")
```

```python
import random

stock = 800      # 初始库存
days = 0        # 经营天数
restock_times = 0  # 补货次数
capacity = 1000  # 仓库容量

while stock <= capacity:
    days += 1
    # 随机生成当天销量
    sales = random.randint(80, 120)
    
    # 当天销售
    if sales > stock:
        sales = stock  # 实际销量不能超过库存
    stock -= sales
    
    # 判断是否需要补货
    if stock < 200:
        stock += 300
        restock_times += 1
    
    print(f"第{days}天 销量:{sales} 库存:{stock}")

print(f"\n连续经营{days}天")
print(f"共补货{restock_times}次")
print(f"最终库存{stock}件")
```

**题目**：投资收益率分析。编写程序分析投资组合，实现：
- 初始资金100万
- 每年收益率在-10%到+20%之间随机波动
- 使用while循环模拟投资过程
- 当资金首次达到150万或低于80万时停止
- 输出投资年限和年度收益率记录

```python
import random

initial = 1000000  # 初始资金
current = initial  # 当前资金
years = 0         # 投资年限
rates = []        # 记录收益率
```

**参考答案**：

```python
import random

initial = 1000000  # 初始资金
current = initial  # 当前资金
years = 0         # 投资年限
rates = []        # 记录收益率

while 800000 <= current <= 1500000:
    years += 1
    rate = random.uniform(-0.1, 0.2)  # 随机收益率
    rates.append(rate)
    
    current = current * (1 + rate)
    print(f"第{years}年 收益率:{rate:.1%} 资金:{current:.2f}")

print(f"\n投资时长：{years}年")
print(f"最终资金：{current:.2f}")
print("历年收益率：")
for i, rate in enumerate(rates, 1):
    print(f"第{i}年：{rate:.1%}")
```

- 使用循环检查每个账户的还款情况
- 出现连续3个月未还款时立即停止检查并发出警告
- 输出风险账户的编号和具体位置

```python
accounts = [
   [1,1,1,1,0,0,1],     # 账户1
   [1,1,0,0,0,1,1],     # 账户2
   [1,1,1,0,1,1,1]      # 账户3
]
```

**参考答案**：

```python
accounts = [
   [1,1,1,1,0,0,1],     # 账户1
   [1,1,0,0,0,1,1],     # 账户2
   [1,1,1,0,1,1,1]      # 账户3
]

risk_found = False
for acc_no, records in enumerate(accounts, 1):
    unpaid_count = 0
   
    for month, paid in enumerate(records, 1):
        if paid == 0:
            unpaid_count += 1
        else:
            unpaid_count = 0
           
        if unpaid_count >= 3:
            print(f"警告：账户{acc_no}在第{month-2}至{month}月连续未还款")
            risk_found = True
            break
           
    if risk_found:
        break

if not risk_found:
    print("未发现高风险账户")
```

**题目**：商品促销活动。模拟一个限时抢购活动，编写程序实现：
- 商品库存100件，每个用户限购1件
- 使用while True模拟用户下单过程
- 每次随机生成下单数量(1-3件)
- 当库存不足或超过限购时使用break终止
- 统计成功订单数和剩余库存

```python
import random

stock = 100          # 初始库存
orders = 0          # 成功订单数
user_purchased = {} # 记录用户购买情况
```

**参考答案**：

```python
import random

stock = 100          # 初始库存
orders = 0          # 成功订单数
user_purchased = {} # 记录用户购买情况

while True:
    user_id = random.randint(1, 200)  # 模拟用户ID
    quantity = random.randint(1, 3)    # 随机下单数量
    
    # 检查是否已购买
    if user_id in user_purchased:
        print(f"用户{user_id}已购买过，订单取消")
        continue
        
    # 检查库存
    if stock < quantity:
        print("库存不足，活动结束")
        break
        
    # 限购检查
    if quantity > 1:
        print(f"用户{user_id}超出限购数量，订单取消")
        continue
        
    # 下单成功
    stock -= quantity
    orders += 1
    user_purchased[user_id] = quantity
    print(f"用户{user_id}下单{quantity}件，剩余库存{stock}件")

print(f"\n活动结束")
print(f"成功订单：{orders}笔")
print(f"剩余库存：{stock}件")
```

- 使用循环模拟交易，当价格触及止损或止盈点时立即卖出
- 输出交易结果和持有天数

```python
prices = [100, 102, 98, 103, 95, 93, 97, 105, 108, 112]
buy_price = prices[0]
buy_day = 1
stop_loss = buy_price * 0.9
stop_profit = buy_price * 1.1
```

**参考答案**：

```python
prices = [100, 102, 98, 103, 95, 93, 97, 105, 108, 112]
buy_price = prices[0]
buy_day = 1
stop_loss = buy_price * 0.9
stop_profit = buy_price * 1.1

for day, price in enumerate(prices[1:], 2):
    print(f"第{day}天 价格:{price}")
    
    if price <= stop_loss:
        profit = price - buy_price
        print(f"\n触发止损：{price:.2f}")
        print(f"交易结果：{profit:.2f}")
        print(f"持有天数：{day-buy_day}天")
        break
        
    if price >= stop_profit:
        profit = price - buy_price
        print(f"\n触发止盈：{price:.2f}")
        print(f"交易结果：{profit:.2f}")
        print(f"持有天数：{day-buy_day}天")
        break
```

**参考答案**：

```python
orders = [
    (1000, 1),  # (金额, 状态)
    (2000, 0),  # 无效订单
    (1500, 1),
    (3000, 0),  # 无效订单
    (2500, 1)
]

valid_count = 0
total_amount = 0

for amount, status in orders:
    if status == 0:
        print(f"跳过无效订单，金额：{amount}")
        continue
        
    valid_count += 1
    total_amount += amount
    print(f"处理有效订单，金额：{amount}")

average = total_amount / valid_count if valid_count > 0 else 0
print(f"\n统计结果：")
print(f"有效订单数：{valid_count}")
print(f"总金额：{total_amount}")
print(f"平均金额：{average:.2f}")
```

```python
scores = [85, -1, 92, 78, -2, 90, 83, -1, 75, 88]

student_count = 0
total_score = 0
excellent_count = 0
```

**参考答案**：

```python
scores = [85, -1, 92, 78, -2, 90, 83, -1, 75, 88]

student_count = 0
total_score = 0
excellent_count = 0

for score in scores:
    if score < 0:  # 特殊情况
        if score == -1:
            print("跳过缺考学生")
        else:
            print("跳过作弊学生")
        continue
        
    student_count += 1
    total_score += score
    
    if score >= 90:
        excellent_count += 1

average = total_score / student_count if student_count > 0 else 0
print(f"\n统计结果：")
print(f"实考人数：{student_count}")
print(f"平均分：{average:.1f}")
print(f"优秀人数：{excellent_count}")
```

#### 练习题12

**题目**：网店数据分析。编写程序分析店铺一个月的订单数据：
- 每日订单数据包含：日期、订单数、退单数
- 使用for循环处理数据，跳过退单率超过20%的异常日期
- 计算正常营业日的平均订单数
- 统计总营业额（假设每单100元）

```python
# 数据格式：(日期, 订单数, 退单数)
daily_data = [
    (1, 100, 5),
    (2, 80, 20),   # 退单率25%，异常
    (3, 120, 15),
    (4, 90, 8),
    (5, 70, 18),   # 退单率25.7%，异常
    (6, 110, 12)
]

valid_days = 0
total_orders = 0
total_revenue = 0
PRICE_PER_ORDER = 100  # 每单收入
```


**参考答案**：

```python
# 数据格式：(日期, 订单数, 退单数)
daily_data = [
    (1, 100, 5),
    (2, 80, 20),   # 退单率25%，异常
    (3, 120, 15),
    (4, 90, 8),
    (5, 70, 18),   # 退单率25.7%，异常
    (6, 110, 12)
]

valid_days = 0
total_orders = 0
total_revenue = 0
PRICE_PER_ORDER = 100  # 每单收入

for day, orders, returns in daily_data:
    return_rate = returns / orders if orders > 0 else 0
    
    if return_rate > 0.2:  # 退单率>20%
        print(f"第{day}日数据异常，退单率：{return_rate:.1%}")
        continue
        
    valid_days += 1
    total_orders += (orders - returns)
    total_revenue += (orders - returns) * PRICE_PER_ORDER

average_orders = total_orders / valid_days if valid_days > 0 else 0
print(f"\n统计结果：")
print(f"正常营业天数：{valid_days}")
print(f"日均有效订单：{average_orders:.1f}")
print(f"总营业额：{total_revenue}元")
```
