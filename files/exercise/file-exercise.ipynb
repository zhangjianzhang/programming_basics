#### 练习题1
题目：编写一个程序，从sales.txt文件中读取每日销售数据（每行格式为"日期,金额"，如"2024-01-01,1234.56"），计算：
1) 总销售额
2) 平均日销售额
3) 将结果写入新文件sales_report.txt，同时包含原始数据和计算结果

参考答案：
```python
# 创建示例数据文件
with open('sales.txt', 'w', encoding='utf-8') as f:
    f.write('2024-01-01,1234.56\n')
    f.write('2024-01-02,2345.67\n')
    f.write('2024-01-03,3456.78\n')

# 处理销售数据
total_sales = 0
sales_count = 0

with open('sales.txt', 'r', encoding='utf-8') as fin:
    sales_data = fin.readlines()
    
    for line in sales_data:
        date, amount = line.strip().split(',')
        total_sales += float(amount)
        sales_count += 1

avg_sales = total_sales / sales_count

# 写入报告
with open('sales_report.txt', 'w', encoding='utf-8') as fout:
    fout.write('销售数据报告\n')
    fout.write('-' * 20 + '\n\n')
    fout.write('原始数据：\n')
    for line in sales_data:
        fout.write(line)
    fout.write('\n统计结果：\n')
    fout.write(f'总销售额：{total_sales:.2f}\n')
    fout.write(f'平均日销售额：{avg_sales:.2f}\n')
```

#### 练习题2
题目：编写程序实现以下功能：
1) 从grades.txt读取学生成绩数据（格式：学号,姓名,成绩）
2) 按成绩从高到低排序
3) 将排序后的数据写入新文件ranked_grades.txt，包含排名信息
4) 只输出前5名学生信息

参考答案：
```python
# 创建示例数据
with open('grades.txt', 'w', encoding='utf-8') as f:
    f.write('2024001,张三,89\n')
    f.write('2024002,李四,95\n')
    f.write('2024003,王五,87\n')
    f.write('2024004,赵六,92\n')
    f.write('2024005,钱七,90\n')
    f.write('2024006,孙八,85\n')

# 读取并处理数据
students = []
with open('grades.txt', 'r', encoding='utf-8') as f:
    for line in f:
        sid, name, score = line.strip().split(',')
        students.append((sid, name, int(score)))

# 排序
students.sort(key=lambda x: x[2], reverse=True)

# 写入排名结果
with open('ranked_grades.txt', 'w', encoding='utf-8') as f:
    f.write('排名,学号,姓名,成绩\n')
    for rank, (sid, name, score) in enumerate(students, 1):
        f.write(f'{rank},{sid},{name},{score}\n')
        if rank <= 5:
            print(f'第{rank}名: {name}, 成绩: {score}')
```

#### 练习题3
题目：编写程序完成以下功能：
1) 从inventory.txt读取商品库存数据（格式：商品编号,名称,数量,单价）
2) 查找库存数量低于10的商品
3) 将这些低库存商品信息写入reorder_list.txt，并计算补货到20个所需的成本

参考答案：
```python
# 创建示例数据
with open('inventory.txt', 'w', encoding='utf-8') as f:
    f.write('A001,笔记本,15,5.5\n')
    f.write('A002,文件夹,8,12.0\n')
    f.write('A003,订书机,5,15.5\n')
    f.write('A004,墨水笔,20,8.5\n')
    f.write('A005,橡皮擦,7,2.0\n')

# 处理库存数据
low_stock_items = []
target_stock = 20

with open('inventory.txt', 'r', encoding='utf-8') as f:
    for line in f:
        item_id, name, quantity, price = line.strip().split(',')
        quantity = int(quantity)
        price = float(price)
        
        if quantity < 10:
            needed = target_stock - quantity
            cost = needed * price
            low_stock_items.append((item_id, name, quantity, price, needed, cost))

# 写入补货清单
with open('reorder_list.txt', 'w', encoding='utf-8') as f:
    f.write('低库存商品补货清单\n')
    f.write('-' * 30 + '\n\n')
    f.write('商品编号,名称,当前库存,单价,需补货数量,补货成本\n')
    total_cost = 0
    
    for item in low_stock_items:
        f.write(f'{item[0]},{item[1]},{item[2]},{item[3]},{item[4]},{item[5]:.2f}\n')
        total_cost += item[5]
    
    f.write(f'\n总补货成本: {total_cost:.2f}元\n')
```

1) 分别统计收入和支出总额
2) 计算当日余额（收入-支出）
3) 将统计结果写入daily_report.txt，包含具体数据和分析结果

参考答案：
```python
# 测试数据准备
with open('transactions.txt', 'w', encoding='utf-8') as f:
    f.write('09:15,收入,1280.50\n')
    f.write('10:30,支出,450.75\n')
    f.write('13:20,收入,2100.00\n')
    f.write('15:45,支出,860.25\n')
    f.write('16:30,收入,1500.00\n')

# 数据处理
total_income = 0
total_expense = 0

with open('transactions.txt', 'r', encoding='utf-8') as f:
    for line in f:
        time, type_, amount = line.strip().split(',')
        amount = float(amount)
        if type_ == '收入':
            total_income += amount
        else:
            total_expense += amount

balance = total_income - total_expense

# 生成报告
with open('daily_report.txt', 'w', encoding='utf-8') as f:
    f.write('每日交易报告\n')
    f.write('=' * 30 + '\n\n')
    f.write(f'总收入：{total_income:,.2f}元\n')
    f.write(f'总支出：{total_expense:,.2f}元\n')
    f.write(f'当日余额：{balance:,.2f}元\n')
```

2) 找出消费金额最高的前3名客户
3) 将结果写入vip_report.txt文件

```python
# 测试数据准备
with open('customer_data.txt', 'w', encoding='utf-8') as f:
    f.write('C001,手机,5999.00,1\n')
    f.write('C002,耳机,299.00,2\n')
    f.write('C001,平板,2999.00,1\n')
    f.write('C003,笔记本,6999.00,1\n')
    f.write('C002,充电器,99.00,3\n')

# 处理数据
customer_totals = {}

with open('customer_data.txt', 'r', encoding='utf-8') as f:
    for line in f:
        cid, product, price, quantity = line.strip().split(',')
        amount = float(price) * int(quantity)
        customer_totals[cid] = customer_totals.get(cid, 0) + amount

# 排序并生成报告
top_customers = sorted(customer_totals.items(), key=lambda x: x[1], reverse=True)

with open('vip_report.txt', 'w', encoding='utf-8') as f:
    f.write('VIP客户消费报告\n')
    f.write('=' * 30 + '\n\n')
    f.write('客户ID    消费金额\n')
    
    for i, (cid, total) in enumerate(top_customers[:3], 1):
        f.write(f'{cid}    {total:,.2f}元\n')
```

1) 让用户输入一个人民币金额
2) 计算可兑换的各种外币数量
3) 将结果写入conversion_result.txt文件

参考答案：
```python
# 测试数据准备
with open('exchange_rates.txt', 'w', encoding='utf-8') as f:
    f.write('USD,7.1842\n')
    f.write('EUR,7.8601\n')
    f.write('JPY,0.0483\n')
    f.write('GBP,9.1245\n')
    f.write('AUD,4.7523\n')

# 获取用户输入
rmb_amount = float(input('请输入人民币金额：'))

# 计算并生成报告
with open('exchange_rates.txt', 'r', encoding='utf-8') as fin:
    rates = {}
    for line in fin:
        currency, rate = line.strip().split(',')
        rates[currency] = float(rate)

with open('conversion_result.txt', 'w', encoding='utf-8') as fout:
    fout.write(f'人民币 {rmb_amount:,.2f}元 的外币兑换结果\n')
    fout.write('=' * 40 + '\n\n')
    
    for currency, rate in rates.items():
        foreign_amount = rmb_amount / rate
        fout.write(f'{currency}: {foreign_amount:,.2f}\n')
```

#### 练习题7
题目：假设您在数据分析部门工作，需要处理daily_prices.txt文件中记录的股票每日价格数据。文件每行格式为"日期,开盘价,收盘价"（如"2024-01-15,45.67,46.89"）。请编写程序实现：
1) 计算每日涨跌幅（涨跌幅=(收盘价-开盘价)/开盘价*100%）
2) 找出涨幅最大和跌幅最大的日期及数据
3) 将分析结果写入price_analysis.txt文件

参考答案：
```python
# 创建示例数据
with open('daily_prices.txt', 'w', encoding='utf-8') as f:
    f.write('2024-01-15,45.67,46.89\n')
    f.write('2024-01-16,46.90,45.78\n')
    f.write('2024-01-17,45.80,47.25\n')
    f.write('2024-01-18,47.30,46.50\n')
    f.write('2024-01-19,46.55,48.10\n')

# 分析数据
price_changes = []
with open('daily_prices.txt', 'r', encoding='utf-8') as f:
    for line in f:
        date, open_price, close_price = line.strip().split(',')
        open_price = float(open_price)
        close_price = float(close_price)
        change_rate = (close_price - open_price) / open_price * 100
        price_changes.append((date, change_rate, open_price, close_price))

max_up = max(price_changes, key=lambda x: x[1])
max_down = min(price_changes, key=lambda x: x[1])

# 写入分析结果
with open('price_analysis.txt', 'w', encoding='utf-8') as f:
    f.write('股票价格分析报告\n')
    f.write('=' * 30 + '\n\n')
    f.write('每日涨跌情况：\n')
    for date, change, open_p, close_p in price_changes:
        f.write(f'{date}: {change:+.2f}%  开盘:{open_p:.2f}  收盘:{close_p:.2f}\n')
    
    f.write(f'\n最大涨幅：\n')
    f.write(f'日期：{max_up[0]}  涨幅：{max_up[1]:+.2f}%\n')
    f.write(f'\n最大跌幅：\n')
    f.write(f'日期：{max_down[0]}  跌幅：{max_down[1]:+.2f}%\n')
```

题目：编写程序处理一个账户流水账文件account_flow.txt，每行格式为"日期,时间,类型,金额,备注"（如"2024-01-20,09:30,支出,99.90,购物"）。实现以下功能：
1) 按照不同类型（收入/支出）分类统计
2) 计算每种类型的总笔数和总金额
3) 将统计结果写入flow_statistics.txt文件

参考答案：
```python
# 创建示例数据
with open('account_flow.txt', 'w', encoding='utf-8') as f:
    f.write('2024-01-20,09:30,支出,99.90,购物\n')
    f.write('2024-01-20,10:15,收入,5000.00,工资\n')
    f.write('2024-01-20,14:20,支出,35.50,午餐\n')
    f.write('2024-01-20,16:45,支出,150.00,交通\n')
    f.write('2024-01-20,17:30,收入,200.00,报销\n')

# 处理数据
stats = {
    '收入': {'count': 0, 'total': 0.0, 'records': []},
    '支出': {'count': 0, 'total': 0.0, 'records': []}
}

with open('account_flow.txt', 'r', encoding='utf-8') as f:
    for line in f:
        date, time, type_, amount, note = line.strip().split(',')
        amount = float(amount)
        stats[type_]['count'] += 1
        stats[type_]['total'] += amount
        stats[type_]['records'].append((date, time, amount, note))

# 写入统计结果
with open('flow_statistics.txt', 'w', encoding='utf-8') as f:
    f.write('账户流水统计报告\n')
    f.write('=' * 40 + '\n\n')
    
    for type_ in ['收入', '支出']:
        f.write(f'{type_}统计：\n')
        f.write(f'总笔数：{stats[type_]["count"]}笔\n')
        f.write(f'总金额：{stats[type_]["total"]:,.2f}元\n')
        f.write('\n明细：\n')
        for date, time, amount, note in stats[type_]['records']:
            f.write(f'{date} {time} {amount:>8,.2f} {note}\n')
        f.write('\n' + '-' * 40 + '\n\n')
```

#### 练习题1
题目：已有一个名为"grades.txt"的文本文件，每行包含一个学生的姓名和4门课程成绩(以空格分隔)，格式如下:
```
张三 88 76 92 85
李四 92 88 76 90
王五 78 92 88 85
```
请编写程序完成以下要求：
1. 定义一个函数calculate_avg(scores)计算一个学生的平均成绩
2. 定义一个函数process_file(filename)读取成绩文件,计算并输出每个学生的平均分
3. 将计算结果写入新文件"results.txt",格式为:姓名 平均分

参考答案：
```python
def calculate_avg(scores):
    """计算平均分"""
    return sum(scores) / len(scores)

def process_file(filename):
    """处理成绩文件"""
    with open(filename, 'r', encoding='utf-8') as f_in:
        with open('results.txt', 'w', encoding='utf-8') as f_out:
            for line in f_in:
                data = line.strip().split()
                name = data[0]
                scores = [float(x) for x in data[1:]]
                avg = calculate_avg(scores)
                print(f"{name}的平均分是: {avg:.2f}")
                f_out.write(f"{name} {avg:.2f}\n")

process_file('grades.txt')
```

#### 练习题2
题目：有一个Web服务器日志文件"server.log",每行记录一次访问,格式如下:
```
192.168.1.1 2023-12-01 10:30:25 /index.html
10.0.0.2 2023-12-01 10:31:12 /about.html
192.168.1.1 2023-12-01 10:32:45 /contact.html
```
请编写程序完成：
1. 定义函数parse_log_line(line)解析一行日志
2. 定义函数analyze_logs(filename)统计每个IP地址的访问次数
3. 将统计结果按访问次数降序写入"stats.txt"

参考答案：
```python
def parse_log_line(line):
    """解析日志行"""
    parts = line.strip().split()
    return {
        'ip': parts[0],
        'date': parts[1],
        'time': parts[2],
        'url': parts[3]
    }

def analyze_logs(filename):
    """分析日志文件"""
    ip_counts = {}
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            log_data = parse_log_line(line)
            ip = log_data['ip']
            ip_counts[ip] = ip_counts.get(ip, 0) + 1
    
    with open('stats.txt', 'w', encoding='utf-8') as f:
        sorted_ips = sorted(ip_counts.items(), 
                          key=lambda x: x[1], 
                          reverse=True)
        for ip, count in sorted_ips:
            f.write(f"{ip}: {count}\n")

analyze_logs('server.log')
```

#### 练习题3
题目：有一个CSV格式的商品数据文件"products.csv",内容如下:
```
商品ID,名称,原价,折扣
001,笔记本电脑,6999,0.8
002,手机,2999,0.9
003,耳机,899,0.7
```
请编写程序完成：
1. 定义函数calculate_price(original, discount)计算折后价
2. 定义函数convert_csv(input_file, output_file)读取CSV文件并转换
3. 生成新的CSV文件,添加一列"折后价"

参考答案：
```python
def calculate_price(original, discount):
    """计算折后价"""
    return float(original) * float(discount)

def convert_csv(input_file, output_file):
    """转换CSV文件"""
    with open(input_file, 'r', encoding='utf-8') as f_in:
        with open(output_file, 'w', encoding='utf-8') as f_out:
            f_out.write('商品ID,名称,原价,折扣,折后价\n')
            next(f_in)
            for line in f_in:
                id, name, price, discount = line.strip().split(',')
                final_price = calculate_price(price, discount)
                f_out.write(f'{id},{name},{price},{discount},'
                          f'{final_price:.2f}\n')

convert_csv('products.csv', 'products_with_final_price.csv')
```

#### 练习题4
题目：某公司有一个名为"sales.txt"的销售数据文件，每行记录一笔销售，格式为"销售日期 产品ID 销售金额"(以空格分隔)，示例数据如下：
```
2023-12-01 P001 2580.5
2023-12-01 P002 1350.0
2023-12-02 P001 3200.0
2023-12-02 P003 980.5
```
请编写程序完成：
1. 定义函数calculate_daily_sales(sales_list)计算某一天的总销售额
2. 定义函数analyze_sales(filename)读取销售数据文件，统计每天的销售总额
3. 将统计结果写入新文件"daily_summary.txt"，格式为：日期 销售总额

参考答案：
```python
def calculate_daily_sales(sales_list):
    """计算一天的销售总额"""
    return sum(float(sale[2]) for sale in sales_list)

def analyze_sales(filename):
    """分析销售数据文件"""
    daily_sales = {}
    
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            date, pid, amount = line.strip().split()
            if date not in daily_sales:
                daily_sales[date] = []
            daily_sales[date].append((date, pid, amount))
    
    with open('daily_summary.txt', 'w', encoding='utf-8') as f:
        for date in sorted(daily_sales.keys()):
            total = calculate_daily_sales(daily_sales[date])
            f.write(f"{date} {total:.2f}\n")
            print(f"{date}的销售总额：{total:.2f}元")

analyze_sales('sales.txt')
```

请编写程序完成：

3. 生成分析报告"expense_report.txt"，包含各部门支出总额和占比

参考答案：
```python
def calculate_department_expenses(expenses_list):
    """计算部门总支出"""
    total = sum(float(expense[3]) for expense in expenses_list)
    return total

def analyze_expenses(filename):
    """分析费用支出"""
    department_expenses = {}
    total_expense = 0
    
    with open(filename, 'r', encoding='utf-8') as f:
        next(f)  # 跳过表头
        for line in f:
            date, dept, type, amount = line.strip().split(',')
            if dept not in department_expenses:
                department_expenses[dept] = []
            department_expenses[dept].append((date, dept, type, amount))
            total_expense += float(amount)
    
    with open('expense_report.txt', 'w', encoding='utf-8') as f:
        f.write("部门费用支出分析报告\n")
        f.write("-" * 30 + "\n\n")
        
        for dept in sorted(department_expenses.keys()):
            dept_total = calculate_department_expenses(department_expenses[dept])
            percentage = (dept_total / total_expense) * 100
            
            f.write(f"部门：{dept}\n")
            f.write(f"总支出：{dept_total:.2f}元\n")
            f.write(f"占比：{percentage:.2f}%\n\n")

analyze_expenses('expenses.csv')
```

#### 练习题6
题目：一个股票交易数据文件"stock_trades.txt"，每行记录一笔交易，格式为：日期 股票代码 交易类型(buy/sell) 数量 价格，示例数据：
```
2023-12-01 600001 buy 100 12.50
2023-12-01 600002 sell 200 8.75
2023-12-02 600001 sell 50 13.20
2023-12-02 600003 buy 300 5.60
```
请编写程序完成：
1. 定义函数calculate_transaction(type, quantity, price)计算交易金额
2. 定义函数analyze_trades(filename)分析每支股票的交易情况
3. 生成报告"trade_summary.txt"，包含每支股票的交易总额和交易次数

参考答案：
```python
def calculate_transaction(type, quantity, price):
    """计算交易金额"""
    amount = float(quantity) * float(price)
    return amount if type == "buy" else -amount

def analyze_trades(filename):
    """分析交易数据"""
    stock_stats = {}  # 存储股票统计信息
    
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            date, code, type, qty, price = line.strip().split()
            amount = calculate_transaction(type, qty, price)
            
            if code not in stock_stats:
                stock_stats[code] = {'total': 0, 'count': 0}
            
            stock_stats[code]['total'] += amount
            stock_stats[code]['count'] += 1
    
    with open('trade_summary.txt', 'w', encoding='utf-8') as f:
        for code in sorted(stock_stats.keys()):
            stats = stock_stats[code]
            f.write(f"股票代码：{code}\n")
            f.write(f"净交易额：{stats['total']:.2f}\n")
            f.write(f"交易次数：{stats['count']}\n\n")

# 调用主函数
analyze_trades('stock_trades.txt')
```

#### 练习题7
题目：一个商品销售数据文件"sales_data.csv"，包含商品销售记录，格式如下：
```
日期,商品代码,单价,数量,折扣
2023-12-01,A001,299.00,2,0.8
2023-12-01,B002,199.50,3,0.9
2023-12-02,A001,299.00,1,1.0
2023-12-02,C003,499.00,2,0.7
```
请编写程序完成：
1. 定义函数calculate_item_revenue(price, quantity, discount)计算单项实际销售额
2. 定义函数process_sales(filename)处理销售数据文件
3. 生成"revenue_report.txt"，统计每个商品的总销售额、销售次数和平均折扣

参考答案：
```python
def calculate_item_revenue(price, quantity, discount):
    """计算实际销售额"""
    return float(price) * int(quantity) * float(discount)

def process_sales(filename):
    """处理销售数据"""
    products = {}
    
    with open(filename, 'r', encoding='utf-8') as f:
        next(f)  # 跳过表头
        for line in f:
            date, code, price, qty, disc = line.strip().split(',')
            revenue = calculate_item_revenue(price, qty, disc)
            
            if code not in products:
                products[code] = {
                    'revenue': 0,
                    'count': 0,
                    'discounts': []
                }
            
            products[code]['revenue'] += revenue
            products[code]['count'] += 1
            products[code]['discounts'].append(float(disc))
    
    with open('revenue_report.txt', 'w', encoding='utf-8') as f:
        for code in sorted(products.keys()):
            stats = products[code]
            avg_discount = sum(stats['discounts']) / len(stats['discounts'])
            
            f.write(f"商品代码：{code}\n")
            f.write(f"总销售额：{stats['revenue']:.2f}\n")
            f.write(f"销售次数：{stats['count']}\n")
            f.write(f"平均折扣：{avg_discount:.2f}\n\n")

# 调用主函数
process_sales('sales_data.csv')
```

#### 练习题8
题目：金融投资部门的股票投资组合文件"portfolio.csv"，记录了投资组合的构成，格式如下：
```
股票代码,持仓数量,买入均价,当前价格
600001,1000,15.50,16.20
600002,2000,8.80,8.60
600003,1500,22.30,23.15
600004,800,45.60,44.80
```
请编写程序完成：
1. 定义函数calculate_position_value(quantity, price)计算持仓市值
2. 定义函数analyze_portfolio(filename)分析投资组合状况
3. 生成"portfolio_analysis.txt"，显示每支股票的市值、盈亏额、盈亏比例，并计算组合总值

参考答案：
```python
def calculate_position_value(quantity, price):
    """计算持仓市值"""
    return int(quantity) * float(price)

def analyze_portfolio(filename):
    """分析投资组合"""
    total_portfolio_value = 0
    
    with open(filename, 'r', encoding='utf-8') as f:
        next(f)  # 跳过表头
        with open('portfolio_analysis.txt', 'w', encoding='utf-8') as out:
            out.write("投资组合分析报告\n")
            out.write("-" * 40 + "\n\n")
            
            for line in f:
                code, qty, cost, price = line.strip().split(',')
                
                # 计算各项指标
                current_value = calculate_position_value(qty, price)
                cost_basis = calculate_position_value(qty, cost)
                profit_loss = current_value - cost_basis
                profit_pct = (profit_loss / cost_basis) * 100
                
                # 累计组合总值
                total_portfolio_value += current_value
                
                # 写入个股分析
                out.write(f"股票代码：{code}\n")
                out.write(f"持仓市值：{current_value:.2f}\n")
                out.write(f"盈亏金额：{profit_loss:.2f}\n")
                out.write(f"盈亏比例：{profit_pct:.2f}%\n\n")
            
            # 写入组合总值
            out.write("-" * 40 + "\n")
            out.write(f"投资组合总市值：{total_portfolio_value:.2f}\n")

analyze_portfolio('portfolio.csv')
```

#### 练习题9
题目：国际贸易公司的订单数据文件"orders.txt"，记录了各国客户的订单信息，每行格式为：订单号 国家 货物重量(kg) 运费单价(USD/kg)，示例数据如下：
```
ORD001 Japan 520.5 2.8
ORD002 Korea 180.0 2.5
ORD003 Japan 320.8 2.8
ORD004 Vietnam 420.0 2.1
```
请编写程序完成：
1. 定义函数calculate_shipping(weight, price)计算订单运费
2. 定义函数analyze_orders(filename)统计各国订单数据
3. 将分析结果写入"shipping_summary.txt"，包含各国的订单数量、总重量、总运费

参考答案：
```python
def calculate_shipping(weight, price):
    """计算订单运费"""
    return float(weight) * float(price)

def analyze_orders(filename):
    """分析订单数据"""
    country_stats = {}
    
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            order_id, country, weight, price = line.strip().split()
            shipping_cost = calculate_shipping(weight, price)
            
            if country not in country_stats:
                country_stats[country] = {
                    'orders': 0,
                    'total_weight': 0.0,
                    'total_cost': 0.0
                }
            
            stats = country_stats[country]
            stats['orders'] += 1
            stats['total_weight'] += float(weight)
            stats['total_cost'] += shipping_cost
    
    with open('shipping_summary.txt', 'w', encoding='utf-8') as f:
        for country, stats in sorted(country_stats.items()):
            f.write(f"国家：{country}\n")
            f.write(f"订单数量：{stats['orders']}\n")
            f.write(f"总重量：{stats['total_weight']:.1f} kg\n")
            f.write(f"总运费：{stats['total_cost']:.2f} USD\n\n")

analyze_orders('orders.txt')
```
