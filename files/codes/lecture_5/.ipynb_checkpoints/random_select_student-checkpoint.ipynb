### 2. 随机生成200个学号

```python
import random
random.seed(100) # 设定random的随机数种子，确保每次生成的学号是相同的
```

```python
# 生成200个不重复的学号
stu_id_list = []
while len(stu_id_list) != 200:
    random_id = random.randint(20210000,20219999)
    if random_id not in stu_id_list:
        stu_id_list.append(random_id)
```

```python
stu_id_list
```

```python
student_list = list(zip(stu_id_list, names_list))
```

```python
student_list
```
