```python
# 全部变为小写
# 去掉两端空白
# 用空白将字符串进行分割
# 最终结果为词列表
word_list = cet4_text.lower().strip().split()
```

```python
# 看看词列表的长度
len(word_list)
```

```python
# 用一个字典来保存单词和对应的词频
# key是单词，value是词频
word_dict = {}
```

```python
# 使用for循环遍历词表
# 使用if条件判断过滤掉不符合要求的单词
# 过滤条件有三个：
#（1）只能包含英文字母；
#（2）单词不能在停用词表中；
#（3）单词长度要大于2，不大于2的单词基本是介词、冠词和单个字母，没什么意义
for word in word_list:
    if (word.encode().isalpha() and
        word not in stopwords_list and
        len(word) > 2):
        word_dict[word] = word_dict.get(word,0)
        word_dict[word] += 1
```

```python
# 看看统计完词频后词表的规模
len(word_dict)
```

```python
# 使用sorted函数对单词字典按照词频进行降序排序
word_freq = sorted(word_dict.items(), key = lambda item:item[1], reverse=True)
```

```python
# 从排序结果中取前300个即可
freq_300 = word_freq[:300]
```

```python
# 把结果都打印出来，大概看一眼，万一有不合适的，就去修改前面代码
for word, freq in freq_300:
    print(word, freq)
```

<b><font color=red>注意</font></b>：

1. `'考'.isalpha()`的结果为`True`，为了解决这种中文判断错误问题，可以将字符串编码后再做判断，`'考'.encode().isalpha()`；

2. 从结果中可以看到，不进行词形还原（lemmatization）会对结果带来噪音，例如`allows`、`allow`、`allowing`本质上应该对应`allow`一个单词即可，可以使用`nltk`工具包进行对文本进行更好的预处理（分句、分词、词形还原等），从而降低结果中的噪音；

3. 结果中的高频300词好像并不是我们期望中的四级必背高频词，这些词语大家在中学时都已经很熟悉了，但是它们真的很常用，所以词频排名就非常靠前了。如果想找到对备考更有价值的高频词，不妨看看词频排名稍微靠后一点的词语。
